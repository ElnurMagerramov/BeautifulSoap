```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
URL = "https://bina.az/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
houses=soup.find(class_="items_container")
```

```python
houses_array=[]
```

```python
location=houses.find_all(class_="location")
list(map(lambda i:houses_array.append(i.get_text()),location))
print(len(houses_array))
```

```python
link=houses.select(".items-i .item_link")
list(map(lambda i:houses_array.append("https://bina.az/"+i.get("href")),link[0:32]))
print(len(houses_array));
```

```python
date=houses.find_all(class_="city_when")
list(map(lambda i:houses_array.append(i.get_text()),date));
```

```python
price=houses.find_all(class_="price")
list(map(lambda i:houses_array.append(i.get_text()),price));
```

```python
houses_array=np.array(houses_array).reshape(4,32).T
houses_array
```

```python
df=pd.DataFrame(houses_array)
df.rename(columns = {0:'Locations', 1:'Links',2:'Dates', 3:'Prices'}, inplace = True)
df.to_csv("binaaz.csv")
csv=pd.read_csv("binaaz.csv")
csv=csv.drop("Unnamed: 0", axis=1)
csv
```
